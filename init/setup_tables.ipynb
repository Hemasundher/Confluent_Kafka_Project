To Create required Delta Tables

```python
catalog = "sales_catalog"
schema = "main"

spark.sql(f"CREATE CATALOG IF NOT EXISTS {catalog}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog}.{schema}")
```

```python
spark.sql(f"USE CATALOG {catalog}")
spark.sql(f"USE SCHEMA {schema}")
```

```python
from pyspark.sql.types import (
    StructType, StructField,
    StringType, DoubleType,
    IntegerType, TimestampType
)
```

```python
# ── 1. users ──────────────────────────────
users_schema = StructType([
    StructField("user_id",  StringType(),  nullable=False),
    StructField("username", StringType(),  nullable=False),
    StructField("email",    StringType(),  nullable=False),
    StructField("balance",  DoubleType(),  nullable=False)
])

spark.createDataFrame([], users_schema) \
    .write \
    .format("delta") \
    .mode("ignore") \
    .saveAsTable("users")
```

```python
# ── 2. items ──────────────────────────────
items_schema = StructType([
    StructField("item_id",   StringType(), nullable=False),
    StructField("item_name", StringType(), nullable=False),
    StructField("price",     DoubleType(), nullable=False)
])

spark.createDataFrame([], items_schema) \
    .write \
    .format("delta") \
    .mode("ignore") \
    .saveAsTable("items")
```

```python
inventory_schema = StructType([
    StructField("item_id",            StringType(),  nullable=False),
    StructField("quantity_available", IntegerType(), nullable=False)
])

spark.createDataFrame([], inventory_schema) \
    .write \
    .format("delta") \
    .mode("ignore") \
    .saveAsTable("inventory")
```

```python
# ── 4. orders ──────────────────────────────
orders_schema = StructType([
    StructField("order_id",      StringType(),   nullable=False),
    StructField("user_id",       StringType(),   nullable=False),
    StructField("item_id",       StringType(),   nullable=False),
    StructField("item_quantity", IntegerType(),  nullable=False),
    StructField("category",      StringType(),   nullable=False),
    StructField("status",        StringType(),   nullable=False),
    StructField("timestamp",     TimestampType(),nullable=False)
])

spark.createDataFrame([], orders_schema) \
    .write \
    .format("delta") \
    .mode("ignore") \
    .saveAsTable("orders")
```

```python
order_logs_schema = StructType([
    StructField("order_id",           StringType(),    nullable=False),
    StructField("user_id",            StringType(),    nullable=False),
    StructField("item_id",            StringType(),    nullable=False),
    StructField("item_quantity",      IntegerType(),   nullable=False),
    StructField("status",             StringType(),    nullable=False),
    StructField("user_balance_before",DoubleType(),    nullable=False),
    StructField("timestamp",          TimestampType(), nullable=False)
])

spark.createDataFrame([], order_logs_schema) \
    .write \
    .format("delta") \
    .mode("ignore") \
    .saveAsTable("order_logs")
```

```python
inventory_logs_schema = StructType([
    StructField("item_id",         StringType(),    nullable=False),
    StructField("quantity_before", IntegerType(),   nullable=False),
    StructField("quantity_after",  IntegerType(),   nullable=False),
    StructField("timestamp",       TimestampType(), nullable=False)
])

spark.createDataFrame([], inventory_logs_schema) \
    .write \
    .format("delta") \
    .mode("ignore") \
    .saveAsTable("inventory_logs")
```

```python
notifications_schema = StructType([
    StructField("user_id",       StringType(),    nullable=False),
    StructField("username",      StringType(),    nullable=False),
    StructField("item_name",     StringType(),    nullable=False),
    StructField("item_quantity", IntegerType(),   nullable=False),
    StructField("balance_after", DoubleType(),    nullable=False),
    StructField("timestamp",     TimestampType(), nullable=False)
])

spark.createDataFrame([], notifications_schema) \
    .write \
    .format("delta") \
    .mode("ignore") \
    .saveAsTable("notifications")
```

```python
print("Tables Created: users, items, inventory, orders, order_logs, inventory_logs, notifications")
```

```python
# ── 8. balance_events ──────────────────────
balance_events_schema = StructType([
    StructField("user_id",       StringType(),    nullable=False),
    StructField("amount_change", DoubleType(),    nullable=False),
    StructField("order_id",      StringType(),    nullable=False),
    StructField("timestamp",     TimestampType(), nullable=False)
])

spark.createDataFrame([], balance_events_schema) \
    .write \
    .format("delta") \
    .mode("ignore") \
    .saveAsTable("balance_events")

print("✓ balance_events table ready")

# ── 9. inventory_events ────────────────────
inventory_events_schema = StructType([
    StructField("item_id",         StringType(),    nullable=False),
    StructField("quantity_change", IntegerType(),   nullable=False),
    StructField("order_id",        StringType(),    nullable=False),
    StructField("timestamp",       TimestampType(), nullable=False)
])

spark.createDataFrame([], inventory_events_schema) \
    .write \
    .format("delta") \
    .mode("ignore") \
    .saveAsTable("inventory_events")

print("✓ inventory_events table ready")
```

```python
# ── temp_batch (used internally by stream) ──
temp_batch_schema = StructType([
    StructField("order_id",           StringType(),  nullable=False),
    StructField("user_id",            StringType(),  nullable=False),
    StructField("item_id",            StringType(),  nullable=False),
    StructField("item_quantity",      IntegerType(), nullable=False),
    StructField("category",           StringType(),  nullable=False),
    StructField("username",           StringType(),  nullable=True),
    StructField("balance",            DoubleType(),  nullable=True),
    StructField("item_name",          StringType(),  nullable=True),
    StructField("price",              DoubleType(),  nullable=True),
    StructField("quantity_available", IntegerType(), nullable=True),
    StructField("total_cost",         DoubleType(),  nullable=True),
    StructField("status",             StringType(),  nullable=True),
    StructField("balance_after",      DoubleType(),  nullable=True)
])

spark.createDataFrame([], temp_batch_schema) \
    .write \
    .format("delta") \
    .mode("ignore") \
    .saveAsTable("temp_batch")

print("✓ temp_batch table ready")
```
