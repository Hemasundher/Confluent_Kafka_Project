```python
from confluent_kafka import Consumer
import json

consumer = Consumer(consumer_conf)
consumer.subscribe(['topic_0'])

messages = []

print("Reading messages...")
for _ in range(20):  # poll 20 times
    msg = consumer.poll(timeout=2.0) # get message if didnt get wait 2 seconds
    if msg is None:
        continue
    if msg.error():
        print(f"Error: {msg.error()}")
    else:
        data = json.loads(msg.value().decode('utf-8'))
        messages.append(data)
        print(f"Received: {data}")

# main part where it sets the offset that till here messages read
#releases the connection or else broker thinks still consumer is active
consumer.close()

print(f"\nTotal messages received: {len(messages)}")
```
