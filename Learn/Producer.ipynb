```python
from confluent_kafka import Producer
import json
import random
import time

producer = Producer(producer_conf)

def delivery_report(err, msg):
    if err:
        print(f'Message failed: {err}')
    else:
        print(f'Message sent to {msg.topic()} partition {msg.partition()}')

# Send 10 fake orders
for i in range(10):
    order = {
        'order_id': i,
        'product': random.choice(['laptop', 'phone', 'tablet']),
        'amount': round(random.uniform(100, 1000), 2)
    }
    # sends to internall buffer not reached to broker yet
    producer.produce('topic_0', json.dumps(order).encode('utf-8'), callback=delivery_report) 
    
    #moved to broker from buffer and get backs the message from broker to callback function
    producer.poll(0)
    time.sleep(1)
#checks the buffer whether all finished or not
producer.flush()
print("All messages sent!")
```
