```python
bootstrap = dbutils.secrets.get(scope="kafka_secrets", key="bootstrap_server")
api_key = dbutils.secrets.get(scope="kafka_secrets", key="api_key")
api_secret = dbutils.secrets.get(scope="kafka_secrets", key="api_secret")

# Producer config
producer_conf = {
    'bootstrap.servers': bootstrap,
    'security.protocol': 'SASL_SSL',
    'sasl.mechanisms': 'PLAIN',
    'sasl.username': api_key,
    'sasl.password': api_secret
}

# Consumer config
consumer_conf = {
    'bootstrap.servers': bootstrap,
    'security.protocol': 'SASL_SSL',
    'sasl.mechanisms': 'PLAIN',
    'sasl.username': api_key,
    'sasl.password': api_secret,
    'group.id': 'databricks-consumer-group',
    'auto.offset.reset': 'earliest'
}

print("Config loaded successfully!")
```
